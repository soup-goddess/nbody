```python
#==========================================================
#ASTRONOMY COORDINATES AND UNITS
#==========================================================
import astropy.units as u 

#==========================================================
#FUNCTION CALLING 
#==========================================================
from full_simulation_modifiable_library import Galactic_stream
from full_simulation_modifiable_library import GlobularClusterLibrary

#==========================================================
#MATH AND PLOTTING
#==========================================================
import matplotlib.pyplot as plt
```

```python
#==========================================================
# LIBRARY INITIALIZATION
#==========================================================
# Create an instance of your cluster library, it will now load from 'globular_clusters.csv'
cluster_library = GlobularClusterLibrary(data_file='globular_clusters.csv')

# See what clusters are available (these now come from the CSV)
print("Available clusters:", cluster_library.list_clusters())

# Instantiate your galactic stream object
galactic_stream = Galactic_stream()

cluster_library.add_cluster(
    name="NGC 6569", 
    ra=273.412, dec=-31.827 , dis=10.53,                        # ra = Right Ascension in degrees, dec = Declination in degrees, dis = Distance from sun in Kpc
    pm_ra_cosdec=-4.125, pm_dec=-7.354, rv=-49.82, mass=2.3E5, #pm_ra_cosdec = , pm_dec = , rv = Radial Velocity, mass = Mass in solar masses 
    brv = -25, rrv = -75                                     # brv = Blue Shifted Radial Velocity, rrv = Red Shifted Radial Velocity 
)
cluster_library.save_clusters_to_file() 
```

```python
#==========================================================
# DATA INITIALIZATION
#==========================================================
# Select cluster from library
selected_cluster_name = "NGC 6569" 

# Set cluster name to be equal to gc_data throughout code
gc_data = cluster_library.get_cluster_data(selected_cluster_name)

# Call function
galactic_stream = Galactic_stream()

# Print data as to ensure it is correct
print(f"--- Analyzing Globular Cluster: {gc_data.name} ---")
print(f"RA: {gc_data.ra:.2f} deg, Dec: {gc_data.dec:.2f} deg")
print(f"Distance: {gc_data.dis:.2f} kpc, Mass: {gc_data.mass:.0e} Msun")
```

```python
#==========================================================
# INTEGRATE AND DISPLAY ORBITS ON 2D GRAPHS
#==========================================================

# Given radial velocity
orbital_figure = galactic_stream.Orbit_display(gc_data=gc_data, shift_rv=None, dt=-0.5 * u.Myr, color='green',
                                               make_2dplot=True, make_3dplot=False)
# Red-shift radial velocity
orbital_figure_redshift = galactic_stream.Orbit_display(gc_data=gc_data, shift_rv=gc_data.rrv, dt=-0.5 * u.Myr,
                                                        color='red', make_2dplot=True, make_3dplot=False)
# Blue-shift radial velocity
orbital_figure_blueshift = galactic_stream.Orbit_display(gc_data=gc_data, shift_rv=gc_data.brv, dt=-0.5 * u.Myr,
                                                         color='blue', make_2dplot=True, make_3dplot=False)
plt.tight_layout()
plt.show()
```

```python
#==========================================================
# INTEGRATE AND DISPLAY ORBITS ON 3D GRAPHS
#==========================================================

# Set up the graph 
figure_3d = galactic_stream.Orbit_display(gc_data=gc_data, shift_rv=None, dt=-1 * u.Myr, color='cyan',
                                          make_2dplot=False, make_3dplot=True)
```

```python
#==========================================================
# INTEGRATE AND DISPLAY RA VS RV STREAM
#==========================================================

stream_orbit, coord_df, stream_galactic, stream_icrs, progenitor_pos0 = (
    galactic_stream.stream_projection_data(gc_data=gc_data, dt=-.5 * u.Myr, num_steps=2800)
)
# Set up plot
stream_figure = galactic_stream.generate_ra_rv_stream(
    gc_data=gc_data,
    stream_icrs=stream_icrs,
)

plt.tight_layout()
plt.show()
```

```python
#==========================================================
# INTEGRATE AND DISPLAY RA VS DEC STREAM
#==========================================================

stream_orbit, coord_df, stream_galactic, stream_icrs, progenitor_pos0 = (
    galactic_stream.stream_projection_data(gc_data=gc_data, dt=-0.5 * u.Myr, num_steps=2800)
)
# Set up plot
stream_figure = galactic_stream.generate_ra_dec_stream(
    gc_data=gc_data,
    stream_icrs = stream_icrs,
)

plt.tight_layout()
plt.show()
```

```python
#==========================================================
# INTEGRATE AND DISPLAY INTERACTIVE 3D PLOT
#==========================================================

tidal_tails_fig = galactic_stream.stream_plot_3d(
    gc_data=gc_data,
    stream_orbit = stream_orbit,
)
tidal_tails_fig.show()
```
